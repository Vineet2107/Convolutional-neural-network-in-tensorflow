```python
from keybert import KeyBERT
kw_model = KeyBERT()
```

```python
r_phrases=kw_model.extract_keywords(resume_text, keyphrase_ngram_range=(1, 3),
                              use_maxsum=True, nr_candidates=20, top_n=10,  diversity=0.0)
j_phrases=kw_model.extract_keywords(JD_text, keyphrase_ngram_range=(1, 3),
                              use_maxsum=True, nr_candidates=20, top_n=10,  diversity=0.0)
```

```python
r_phrases=[t[0] for t in r_phrases]
j_phrases=[t[0] for t in j_phrases]
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def vect_cos(vect, test_list):
    """ Vectorise text and compute the cosine similarity """
    query_0 = vect.transform([' '.join(vect.get_feature_names())])
    query_1 = vect.transform(test_list)
    cos_sim = cosine_similarity(query_0.A, query_1.A)
    return query_1, np.round(cos_sim.squeeze(), 3)

vectoriser = CountVectorizer().fit(r_phrases)
vectoriser.vocabulary_


JD_skills_vect, JD_skills_cos = vect_cos(vectoriser, [' '.join(j_phrases)])


print('\nThe cosine similarity between resume keywords and JD keywords is {}.'.format(JD_skills_cos))
```

```python
r_text=resume_text.replace('\n',' ');
j_text=JD_text.replace('\n', ' ');
```

```python
r_text=' '.join(r_text.split())
j_text=' '.join(j_text.split())
```

```python
import math
import re
from collections import Counter

WORD = re.compile(r"\w+")


def get_cosine(vec1, vec2):
    intersection = set(vec1.keys()) & set(vec2.keys())
    numerator = sum([vec1[x] * vec2[x] for x in intersection])

    sum1 = sum([vec1[x] ** 2 for x in list(vec1.keys())])
    sum2 = sum([vec2[x] ** 2 for x in list(vec2.keys())])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator


def text_to_vector(text):
    words = WORD.findall(text)
    return Counter(words)




vector1 = text_to_vector(r_text)
vector2 = text_to_vector(j_text)

cosine = get_cosine(vector1, vector2)

print("Cosine:", cosine)
```

```python
import nltk
nltk.download('stopwords')
from pyresparser import ResumeParser

import os

from docx import Document
```

```python
j_data= ResumeParser(JD_file).get_extracted_data()
r_data= ResumeParser(resume_file).get_extracted_data()
```

```python
r_skills=r_data['skills']
j_skills=j_data['skills']
```

```python
vectoriser = CountVectorizer().fit(r_skills)
vectoriser.vocabulary_
JD_skills_vect, JD_skills_cos = vect_cos(vectoriser, [' '.join(j_skills)])

print('The cosine similarity between resume skills and JD skills is {}.'.format(JD_skills_cos))
```

```python
def Union(lst1, lst2):
    final_list = list(set(lst1) | set(lst2))
    return final_list
  
skills=(Union(r_skills, j_skills))

re_skills=[]
for skill in skills:
  if skill in r_skills:
    re_skills.append('YES')
  else :
    re_skills.append('NO')


jd_skills=[]
for skill in skills:
  if skill in j_skills:
    jd_skills.append("YES")
  else :
    jd_skills.append('NO')
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(skills, re_skills, jd_skills)),
               columns =['Skill', 'In Resume', 'In JD'])
```
